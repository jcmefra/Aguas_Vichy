```python
import matplotlib.pyplot as plt
import numpy as np
import pygimli as pg
from pygimli.physics import ert
```

# November 24 ERT 1

```python
data1 = ert.load('Surveys/nov24_pygimli.dat')
print(data1)
```

```python
ax, cb = ert.show(data1)
```

```python
data1["k"] = ert.geometricFactors(data1)
data1["err"] = ert.estimateError(data1, relativeError=0.02, absoluteUError=50e-6)
mgr = ert.ERTManager(data1, verbose=True,sr=False)
```

```python
mgr.invert(lam=30, verbose=True)
```

```python
kw = dict(logScale=True, cMap="Spectral_r", xlabel="x (m)", ylabel="y (m)", cMin=0, cMax=100000)
ax, cb = mgr.showResult(**kw)
ax.grid(True)
ax.set_ylim([1479, ax.get_ylim()[1]])
```
